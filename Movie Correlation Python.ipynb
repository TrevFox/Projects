```python
# Import Libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)  # Adjusts the configuration of the plots

# Read in the dataset
df = pd.read_csv(r'C:\Users\tafox\Downloads\movies.csv')
```

```python
# Quick look at the data

df.head()
```

```python
# Check for missing data with percentage output

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
    
```

```python
# Remove records with null values

df = df.dropna()
```

```python
# Check that records with null values were removed

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
    
```

```python
# Inspect data types for the columns

df.dtypes
```

```python
# Change data type of columns where being float type is unnecessary

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
```

```python
# Check data type changes

df.dtypes
```

```python
# View dataframe

df
```

```python
# Order Records by Gross Revenue descending

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Remove duplicates if there are any for production company

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Checking for suspected correlation between budget and gross
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Film Budget')

plt.ylabel('Gross Earnings')

plt.show
```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
# Look at correlation

df.corr(method='pearson')
```

```python
# High Correlation between budget and gross
```

```python
# Create correlation Matrix

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Next look at suspected correlation between company and gross
# First need to numerize non-numeric columns to check correlation

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
# Create correlation matrix with all columns now that they are numerized

correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# View without heatmap

df_numerized.corr()
```

```python
# Unstack for readability

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# Sort for further readability

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# View the high correlations only

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```

```python
# Votes and Budget have the highest correlation to Gross earnings
# Production Company did not correlate with Gross as expected
```
